```python
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams['figure.facecolor'] = 'white'
import colors
palette = colors.tol_muted(print_swatch=True)
raw_color = palette[0]
cca_color = palette[6]
spkr_color = palette[2]
mic_color = palette[3]
el_color = palette[4]
sh_color = palette[8]
import mne
from gkTools.preprocessing import preprocessing as eeg
import numpy as np
import csv
```

```python
user = 'kfsh'
data_path = f'/home/{user}/data/'
git_path = eeg.get_git_path(user)
subj = 'OP0007'
```

## Panel X: data schematic

```python
start,stop = 430,444
# picks = ['AFz','Fz','FCz','Cz','CPz','Pz','POz','Oz']
picks = ['Fp1','Fp2','AF7','AF8','Cz','PO7','PO8','Oz']

# Get audio
audio_fpath = f'{data_path}eeg/{subj}/{subj}_B1/{subj}_B1_audio.vhdr'
raw_audio = mne.io.read_raw_brainvision(audio_fpath,preload=True,verbose=False)
aud_chs = ['spkr','mic']
audio = raw_audio.get_data(picks=aud_chs)
audio_resp = eeg.butter_highpass_filter(audio.T,10,raw_audio.info['sfreq']).T
# Get raw & filter
raw_fpath = f'{data_path}eeg/{subj}/{subj}_B1/{subj}_B1_downsampled.vhdr'
raw = mne.io.read_raw_brainvision(raw_fpath,preload=True,verbose=False)
raw.set_eeg_reference(['TP9','TP10'],verbose=False)
raw.notch_filter(60,verbose=False)
raw.filter(l_freq=1,h_freq=15,verbose=False)
# Get cca & filter
cca_fpath = f'{data_path}eeg/{subj}/{subj}_B1/{subj}_B1_cca.vhdr'
cca = mne.io.read_raw_brainvision(cca_fpath,preload=True,verbose=False)
cca.filter(l_freq=1,h_freq=15,verbose=False)
# do some numpy sorcery
tmin,tmax = start*128, stop*128
tmin_a, tmax_a = start*25000, stop*25000
eeg_resp = cca.get_data(picks=picks)
eog_resp = raw.get_data(picks=['vEOG'])
emg_resp = raw.get_data(picks=['hEOG'])
spkr_norm = audio_resp[0,tmin_a:tmax_a]/(np.abs(audio_resp[0,tmin_a:tmax_a]).max())
mic_norm = audio_resp[1,tmin_a:tmax_a]/np.abs(audio_resp[1,tmin_a:tmax_a]).max()
eog_norm = eog_resp[:,tmin:tmax]/np.abs(eog_resp[:,tmin:tmax]).max()
emg_norm = emg_resp[:,tmin:tmax]/np.abs(emg_resp[:,tmin:tmax]).max()
eeg_norm = eeg_resp[:,tmin:tmax]/np.abs(eeg_resp[:,tmin:tmax]).max()
fs25000 = np.vstack((mic_norm,spkr_norm))
fs128 = np.vstack((eeg_norm,eog_norm,emg_norm))
```

```python
# plot
mic_start = np.int(432.41*128)-tmin
spkr_start = np.int(438.13*128)-tmin

mic_start_a = np.int(432.41*25000)-tmin_a
spkr_start_a = np.int(438.13*25000)-tmin_a

labels = ['Production','Perception']+picks+['EOG activity','EMG activity']
nrows = len(labels)
fig, ax = plt.subplots(nrows=nrows,ncols=1,sharex=False,sharey=True,figsize=(8,8))
for i,l in enumerate(labels):
    plt.subplot(nrows,1,i+1)
    if l == 'Production':
        plt.plot(fs25000[i,:].T, color='k')
        plt.gca().set_xlim([0,fs25000.shape[1]])
        plt.axvline(mic_start_a,color=mic_color,label='Production start')
        plt.axvline(spkr_start_a,color=spkr_color,label='Perception start')
    elif l == 'Perception':
        plt.plot(fs25000[i,:].T, color='k')
        plt.gca().set_xlim([0,fs25000.shape[1]])
        plt.axvline(mic_start_a,color=mic_color,label='Production start')
        plt.axvline(spkr_start_a,color=spkr_color,label='Perception start')
    else:
        plt.plot(fs128[i-2,:].T,color='k')
        plt.gca().set_xlim([0,fs128.shape[1]])
        plt.axvline(mic_start,color=mic_color,label='Production start')
        plt.axvline(spkr_start,color=spkr_color,label='Perception start')
    plt.axis('off')
    if l in ['Perception','Production','EOG activity','EMG activity']:
        plt.title(l,fontsize="x-large")
    if i == 2:
        plt.title("EEG activity",fontsize="x-large")
plt.tight_layout()
plt.savefig('./pdf/data_schematic.pdf')
```

```python
# separate legend
fig = plt.figure(figsize=(3,3))
plt.hist([0],color=mic_color);
plt.hist([0],color=spkr_color);
plt.gca().set_ylim([0,5])
plt.legend(['Production start','Perception start'],fontsize='x-large')
plt.tight_layout()
plt.savefig('./pdf/schematic_legend.pdf')
```

## Panel X: comparison of raw data before and after ICA/CCA

```python
chs = dict()
nchans=16
tmin,tmax = 1006,1012
tmin,tmax = tmin*128,tmax*128
resp = annotated_raw.get_data(picks=annotated_raw.info['ch_names'][5:nchans+5],start=tmin, stop=tmax)
fig, ax = plt.subplots(nrows=nchans,ncols=1,sharex=False,sharey=True,figsize=(8,8))
for i,ch in enumerate(resp):
    plt.subplot(nchans,1,i+1)
    plt.plot(ch,color='k')
    chs[i] = ch
    plt.gca().set_ylim([chs[i].min()*1.1,chs[i].max()*1.1])
    plt.gca().set_xlim([0,tmax-tmin])
    plt.gca().set_xticks([])
    plt.gca().set_yticks([])
    plt.box(False)
    plt.ylabel(annotated_raw.info['ch_names'][i+5])
plt.tight_layout();
fig.savefig('./pdf/raw.pdf')
```
